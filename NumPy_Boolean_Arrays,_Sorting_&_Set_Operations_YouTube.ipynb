```python
import numpy as np

rng = np.random.default_rng()
arr = rng.standard_normal(100)
arr
```

```python
(arr > 0).sum() #Number of  positive values
```

```python
(arr <=0).sum() # Number of negative values
```

##any() and all()

```python
bools = np.array([False, False, True, False])

bools.any() #True
bools.all() #False
```
