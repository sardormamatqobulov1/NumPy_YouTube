##Pseudorandom Number
Generation

```python
import numpy as np
```

```python
samples = np.random.standard_normal((4,4))
samples
```

```python
#speed comparison
from random import normalvariate
N = 1_000_000
```

```python
%timeit [normalvariate(0,1) for _ in range(N)]
```

```python
%timeit np.random.standard_normal(N)
```

```python
#Generator and Seed
rng = np.random.default_rng(seed=12345)
rng.standard_normal((2,3))
```

```python
#Common methods
rng.integers(0, 10, size=5)
rng.uniform(0,1, size=5)
```

##Universal Functions(ufuncs)

```python
#Unary ufuncs
arr = np.arange(10)
np.sqrt(arr)
np.exp(arr)
```

```python
#Binary ufuncs
x = rng.standard_normal(8)
y = rng.standard_normal(8)

np.maximum(x,y)
```

```python
#Multiple outputs
arr = rng.standard_normal(7) * 5
frac, whole = np.modf(arr)
```

```python
#out parameter
out = np.zeros_like(arr)
np.add(arr, 1, out=out)
```
