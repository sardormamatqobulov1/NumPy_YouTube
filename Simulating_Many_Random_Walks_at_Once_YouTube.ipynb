```python
import numpy as np
rng = np.random.default_rng()
```

##Setup:Number of Walks and Steps

```python
nwalks = 5000
nsteps = 1000
```

##Generate Random Steps(2D)

```python
draws = rng.integers(0,2,size=(nwalks,nsteps))
```

```python
draws
```

##Convert Coin Flips to Steps

```python
steps = np.where(draws > 0, 1, -1)
```

```python
steps
```

##Compute All Walks at Once

```python
walks = steps.cumsum(axis=1)
```

```python
walks
```

##Global Max&Min

```python
walks.max()
```

```python
walks.min()
```

##Who Reached Â±30?

```python
hits30 = (np.abs(walks) >= 30).any(axis=1)
```

```python
hits30.sum()
```

##Crossing Time(Important Concept)

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(axis=1)
```

```python
crossing_times
```

##Average Crossing Time

```python
crossing_times.mean()
```

##Different Step Distribution

```python
draws = 0.25 * rng.standard_normal((nwalks, nsteps))
```
