```python
import numpy as np

arr = np.arange(10)
np.save("some_array", arr)
```

```python
loaded_arr = np.load("some_array.npy")
print(loaded_arr)
```

```python
np.savez("array_archive.npz", a=arr, b=arr)
```

```python
archive = np.load("array_archive.npz")
print(archive["b"])
```

```python
np.savez_compressed("arrays_compressed.npz", a=arr, b=arr)
```
